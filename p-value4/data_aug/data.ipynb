```python
import pandas as pd
from imblearn.over_sampling import SMOTE
from collections import Counter

# Load the dataset
file_path = '../balanced_data1.csv'
df = pd.read_csv(file_path, header=None)

# Assume the last column is the target (DX)
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

# Check original class distribution
print("Original class distribution:", Counter(y))

# Define desired number of samples per class
target_count = 1500

# Use SMOTE to oversample
smote = SMOTE(sampling_strategy={cls: target_count for cls in set(y)})
X_res, y_res = smote.fit_resample(X, y)

# Check new class distribution
print("New class distribution after SMOTE:", Counter(y_res))

# Combine features and labels into a single DataFrame
augmented_df = pd.concat([pd.DataFrame(X_res), pd.Series(y_res, name='DX')], axis=1)

# Save to CSV
augmented_df.to_csv('balanced_data1_augmented.csv', index=False)

print("Augmented dataset saved as 'balanced_data1_augmented.csv'")
```
